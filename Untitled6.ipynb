```
names = ('id,name,city,country,iata,icao,lat,lon,'
         'alt,timezone,dst,tz,type,source').split(',')

airports = pd.read_csv(
    'https://github.com/ipython-books/'
    'cookbook-2nd-data/blob/master/'
    'airports.dat?raw=true',
    header=None,
    names=names,
    index_col=4,
    na_values='\\N')
airports_Aus = airports[airports['country'] ==
                       'Australia']
airports_Aus
```

We construct the list of edges representing our graph, where nodes are airports, and two airports are connected if there exists a route between them (flight network):

```
conection = flights [['From_City', 'To_City']].values
conection
```

Let's create the networkX graph from the conection array

```
l = nx.from_edgelist(conection)
```

```
len(l.nodes()), len(l.edges())
```

Exist 21 Airports in Australia and 25 routes 

```
fig, ax = plt.subplots(1, 1, figsize=(6, 6))
nx.draw_networkx(l, ax=ax, node_size=8,
                 font_size=8, alpha=.8,
                 width=.8)
ax.set_axis_off()
```
